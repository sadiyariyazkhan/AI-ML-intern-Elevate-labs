```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
percentile25=titanic['Fare'].quantile(0.25)
percentile75=titanic['Fare'].quantile(0.75)
iqr=percentile75-percentile25
upper_limit=percentile75+1.5*iqr
lower_limit=percentile25-1.5*iqr
print(upper_limit)
print(lower_limit)
```

```python
titanic[titanic['Fare']>upper_limit]
titanic[titanic['Fare']<lower_limit]
new_titanic=titanic[titanic['Fare']>upper_limit]
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.histplot(titanic['Fare'])

plt.subplot(2,2,2)
sns.boxplot(titanic['Fare'])

plt.subplot(2,2,3)
sns.histplot(new_titanic['Fare'])

plt.subplot(2,2,4)
sns.boxplot(new_titanic['Fare'])
```
